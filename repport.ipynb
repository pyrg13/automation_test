#### I. Données par zone climatique
##### A. Température  

```python
dfs_dict_1 = dict(sorted(dfs_dict_1.items()))
for zone, df in dfs_dict_1.items():
    df[df.columns.drop('date')] = df[df.columns.drop('date')].round(1)
    plt.figure(figsize=(7,5))
    for column in df.columns.drop('date'):
        plt.plot(df['date'], df[column], label=column)
    plt.legend(loc='upper right')
    plt.title(f'Données de température pour la zone {zone}')
    plt.xticks(rotation=30)
    plt.show()
    print(f"Données moyennes de température pour la zone : {zone}")
    display(df)
```

##### B. Hygrométrie  

```python
dfs_dict_2 = dict(sorted(dfs_dict_2.items()))
for zone, df in dfs_dict_2.items():
    df[df.columns.drop('date')] = df[df.columns.drop('date')].round(1)
    plt.figure(figsize=(7,5))
    for column in df.columns.drop('date'):
        plt.plot(df['date'], df[column], label=column)
    plt.legend(loc='upper right')
    plt.title(f'Données d\'hygrométrie pour la zone {zone}')
    plt.xticks(rotation=30)
    plt.show()
    print(f"Données moyennes de l'hygrométrie pour la zone : {zone}")
    display(df)
```

#### II. Données par indicateur
##### A. Température

```python
correspondance_noms_T = {
    'daily_mean_T': 'T°24h',
    'mean_day_T': 'T° jour',
    'mean_night_T': 'T° nuit',
    'mean_LS_T': 'T° lev-sol',
    'mean_M_T': 'T° matin',
    'mean_AM_T': 'T° après-midi'
}

for var, df in dfs_dict_3.items():
    df[df.columns.drop('date')] = df[df.columns.drop('date')].round(1)
    var_readable = correspondance_noms_T.get(var, var)

    plt.figure(figsize=(7,5))
    for column in df.columns.drop('date'):
        plt.plot(df['date'], df[column], label=column)
    plt.legend()
    plt.xticks(rotation=30)
    plt.title(f'Données thermique ({var_readable}) pour toutes les zones')
    plt.show()
    print(f"Données moyennes de {var_readable} pour toutes les zones")
    display(df)
```
